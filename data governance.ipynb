# Task
Visualize the trend of a user-specified indicator over time for a user-specified list of countries using the `df_data` DataFrame.

## Identify the indicator and countries

### Subtask:
Ask the user to specify which indicator and which countries they are interested in visualizing.


**Reasoning**:
Print the unique indicator names to allow the user to choose.



```python
print("Available Indicators:")
for indicator in df_data['Indicator Name'].unique():
    print(indicator)
```

**Reasoning**:
Ask the user for their desired indicator and list of countries and store their input in variables.



```python
selected_indicator = input("Please enter the exact name of the indicator you want to visualize: ")

print("\nAvailable Countries:")
for country in df_data['Country Name'].unique():
    print(country)

selected_countries_input = input("\nPlease enter a comma-separated list of country names you want to visualize: ")

selected_countries = [country.strip() for country in selected_countries_input.split(',')]
```

## Filter the data

### Subtask:
Filter the `df_data` DataFrame to include only the selected indicator and countries.


**Reasoning**:
Filter the DataFrame based on the selected indicator and countries as per the instructions.



```python
df_filtered = df_data[df_data['Indicator Name'] == selected_indicator]
df_filtered = df_filtered[df_filtered['Country Name'].isin(selected_countries)]
display(df_filtered.head())
```

## Reshape the data

### Subtask:
Transform the filtered data from wide format (years as columns) to long format (years as rows) to make it suitable for plotting time series for multiple countries.


**Reasoning**:
Transform the filtered data from wide to long format using melt to prepare it for time series plotting and convert the year column to numeric.



```python
id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']
value_vars = [col for col in df_filtered.columns if col not in id_vars and col != 'Unnamed: 66']

df_long = df_filtered.melt(
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='Year',
    value_name='Value'
)

df_long['Year'] = pd.to_numeric(df_long['Year'])

display(df_long.head())
```

## Visualize the data

### Subtask:
Create a line plot showing the indicator's value over time for each selected country.


**Reasoning**:
Create a line plot to visualize the trend of the selected indicator over time for each selected country.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_long, x='Year', y='Value', hue='Country Name')
plt.title(f'Trend of {selected_indicator} over Time for Selected Countries')
plt.xlabel('Year')
plt.ylabel('Indicator Value')
plt.show()
```

## Refine and display the plot

### Subtask:
Add titles, labels, and legends to the plot for clarity and display it.


**Reasoning**:
Add titles, labels, and legends to the plot for clarity and display it.



```python
plt.title(f'Trend of {selected_indicator} over Time for Selected Countries')
plt.xlabel('Year')
plt.ylabel('Indicator Value')
plt.legend(title='Country')
plt.show()
```

**Reasoning**:
The legend did not appear because the previous plotting command did not include labels for the lines. Adding the 'label' argument to the lineplot will allow the legend to display correctly.



```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_long, x='Year', y='Value', hue='Country Name', label='Country Name')
plt.title(f'Trend of {selected_indicator} over Time for Selected Countries')
plt.xlabel('Year')
plt.ylabel('Indicator Value')
plt.legend(title='Country')
plt.show()
```

**Reasoning**:
The legend still did not appear because the `label` argument in `sns.lineplot` doesn't automatically create labels for the legend when `hue` is used. The legend is automatically generated by `seaborn` when using `hue`, so calling `plt.legend()` manually without specifying handles and labels is causing the warning. The previous `sns.lineplot` call already included the `hue='Country Name'` argument which is sufficient for generating the legend automatically by seaborn. The previous code block was trying to call `plt.legend()` manually after `seaborn` has already generated the legend. Removing the explicit `plt.legend()` call will allow the seaborn-generated legend to be displayed.



```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_long, x='Year', y='Value', hue='Country Name')
plt.title(f'Trend of {selected_indicator} over Time for Selected Countries')
plt.xlabel('Year')
plt.ylabel('Indicator Value')
plt.show()
```
