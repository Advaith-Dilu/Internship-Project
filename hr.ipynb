```python
#encoding +zscore outlier removal
import pandas as pd
df=pd.read_csv("HRDataset_v14.csv")
from scipy.stats import zscore
#drop non numerical for zscore
df_numeric=df.select_dtypes(include=np.number)
z_scores=np.abs(zscore(df_numeric))
#keep rows where all zscores are less than 3
df_cleaned=df[(z_scores<3).all(axis=1)]
print ("shape after removing outliers:",df_cleaned.shape)
```

```python
import pandas as pd
df=pd.read_csv("HRDataset_v14.csv")
#Group based insights
#AVERAGE SALARY BY DEPARTMENT
df.groupby('Department')['Salary'].mean().sort_values(ascending=False)
#count of Employees by performancescore
sns.countplot(x='PerformanceScore',data=df)
plt.title('performance score distribution')
plt.show()
```

```python
#Average salary by  Each department
dept_salary=df.groupby('Department')['Salary'].mean().sort_values(ascending=False)
print(dept_salary)
#plot
plt.figure(figsize=(10,5))
sns.barplot(x=dept_salary.values,y=dept_salary.index,palette='viridis')
plt.title('Average Salary by Department')
plt.xlabel('Average Salary')
plt.ylabel('Department')
plt.show()
```

```python
#employee satisfactiobn by position
satisfaction=df.groupby('Position')['EmpSatisfaction'].mean().sort_values()
#plot
plt.figure(figsize=(10,6))
sns.barplot(x=satisfaction.values,y=satisfaction.index,palette='mako')
plt.title('Average Employee satisfaction by position')
plt.xlabel('Average Satisfaction')
plt.ylabel('Position')
plt.show()
```

```python
#Turn over rate by department
# 'Termd' column (0=active, 1 = left)
turnover = df.groupby('Department')['Termd'].mean().sort_values(ascending=False)*100
print(turnover)
#plot
plt.figure(figsize=(10,5))
sns.barplot(x=turnover.values,y=turnover.index,palette='coolwarm')
plt.title('Turnover Rate by Department(%)')
plt.xlabel('Turnover Rate(%)')
plt.ylabel('Department')
plt.show()
```
