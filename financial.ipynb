```python
if 'Sales' in df.columns and 'Profit' in df.columns:
    sns.scatterplot(data=df, x='Sales', y='Profit', hue='Country' if 'Country' in df.columns else None)
    plt.title('Sales vs Profit')
    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f'Outliers in {col}')
    plt.show()
```

```python
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1
outliers = ((df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))).sum()
print("Outlier counts per column:\n", outliers)
```

```python
if 'Profit' in df.columns:
    corr_target = df.corr(numeric_only=True)['Profit'].sort_values(ascending=False)
    print("Correlation with Profit:\n", corr_target)
```

```python
if 'Country' in df.columns and 'Sales' in df.columns:
    country_sales = df.groupby('Country')['Sales'].sum().sort_values(ascending=False)
    country_sales.plot(kind='bar', figsize=(8,4), title='Total Sales by Country')
    plt.show()
```
